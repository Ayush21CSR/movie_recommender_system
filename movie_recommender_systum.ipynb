```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
# now we have to apply a transformation over the columns=[genres,keywords,cast,crew] and that transformation is:-
# if we have james camroon then convert it to jamescamroon, so we will remove the spaces between the words
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_df=movies[['id','title','tags']]
```

```python
new_df
```

```python
# converting list to string
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head(2)
```

```python
# now we will remove the connecting words(and,to,or,then,at,for,is..... etc) then we will extract the root words (loving=>love,caring=>care,talked=>talk)
#then we will vectorize the text(converting text to vector)
```

```python
# now we will apply stemming
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=6000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
# now in vectors we are not using euclidian distance insted we are using angle between the vectors by cosine_similarity where '0' signify least similarity and '1' signify highest similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
# now we willl make a function in which if we pass a movie then it will ge me top 10 similar movies to it
```

```python
# first we will find the index of the movie given. then from index mai similarity matrix me ghusunga aur us perticular index ki movies nikalunga then i will sort the distances and return top 5 movies
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list= sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
